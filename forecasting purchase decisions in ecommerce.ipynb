## Feature Engineering

```python
df['purchase_price'] = df.apply(lambda row: row['price'] if row['event_type'] == 'purchase' else 0, axis=1)

user_summary = df.groupby('user_id').agg(
    frequency=('event_time', 'count'),
    num_views=('event_type', lambda x: (x == 'view').sum()),
    num_carts=('event_type', lambda x: (x == 'cart').sum()),
    num_purchases=('event_type', lambda x: (x == 'purchase').sum()),
    total_spent=('purchase_price', 'sum')
).reset_index()
```

```python
user_summary
```

```python
user_summary[['frequency', 'num_views', 'num_carts', 'num_purchases', 'total_spent']].describe()
```

Top Spenders

```python
top_spenders = user_summary.sort_values(by='total_spent', ascending=False).head(10)
top_spenders
```

## Classification

```python
user_summary['made_purchase'] = user_summary['num_purchases'].apply(lambda x: 1 if x > 0 else 0)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.metrics import accuracy_score, precision_score
```

```python
features = ['frequency', 'num_views', 'num_carts', 'total_spent']
X = user_summary[features]
y = user_summary['made_purchase']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
cm =confusion_matrix(y_test, y_pred)
print(cm)
```

```python
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
print(classification_report(y_test, y_pred))
```

```python
accuracy= accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)

print(accuracy)
print(precision)
```
